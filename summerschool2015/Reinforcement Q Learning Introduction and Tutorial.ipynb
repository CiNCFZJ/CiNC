# Reward-driven learning and behavior adaptation - Reinforcement Learning Introduction with Q-Learning.

## One state, two actions

$\newcommand{\argmax}[1]{\underset{#1}{\operatorname{argmax}}}
\newcommand{\argmin}[1]{\underset{#1}{\operatorname{argmin}}}$

You have two actions you must choose between.  Each will produce some unknown, stochastic feedback value from your environment.  Which action do you choose?

Right.  Try each one a bunch of times and see which one results in a higher average feedback value.

Let the action choice at time $t$ be $a_t$, and its possible values are $a_t \in \{1,2\}$.  Let the feedback value, which we will call the reward, for action $a_t$ be $r(a_t)$.  We don't know what the possible values of $r(a_t)$ are.

At any time $t$ we can pick what we currently believe to be the optimal action if we had a prediction of the reward for that action.  Define this prediction to be $Q(a_t)$.  Assuming this function exists and is correct, we can greedily pick what we think is the best action by

$$a_t = \argmax{a\in\{1,2\}} Q(a)$$.

So, how can we learn the $Q$ function?  Let's define $Q$ to be a table, or tuple, of two values, one for each action. We can minimize the squared error of $Q$ by simply going down the gradient of $(r(a_t) - Q(a_t))^2$ with respect to $Q(a_t)$ on each step.

$$ Q(a_t) = Q(a_t) + \rho (r(a_t) - Q(a_t))$$

for $\rho > 0$ being a small step size, so that $Q(a_t)$ converges on the expected value (mean) of $r(a_t)$.

Let's try an example.  Define the initial $Q$ function to be $Q=(0,0)$, and the reward function to be
$$r(a) = \begin{cases}
10+z; & a=1\\
20+z; & a=2,
\end{cases}$$

for $z \sim \mathcal{N}(0,1)$.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

Q = [0,0]
def reward(a):
    z = np.random.normal(0,1)
    if a == 1:
        return 10 + z
    else:
        return 20 + z
```

```python
reward(1), reward(1), reward(2), reward(2)
```

```python
Q = [0,0]
rho = 0.2
log = {'Q':[], 'a':[],'r':[]}
for steps in range(100):
    a = np.random.choice((1,2))
    r = reward(a)
    Q[a-1] = Q[a-1] + rho * (r - Q[a-1])
    log['Q'].append(Q[:])
    log['a'].append(a)
    log['r'].append(r)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(3,1,1)
plt.plot(log['a'],'o-')
plt.ylabel("Action Chosen")
plt.subplot(3,1,2)
plt.plot(log['r'])
plt.ylabel("Reward")
plt.subplot(3,1,3)
plt.plot(np.array(log['Q']),lw=3)
plt.xlabel("Step")
plt.ylabel("Q value")
plt.legend(('$Q(a=1)$', '$Q(a=2)$'));
```

We can define a policy function, $\pi(a)$, that is the probability of selecting each action.  The greedy policy of selecting the action that is best given the current $Q$ function, is
$$\pi(a) = \begin{cases}
1, & a = \argmax{b} Q(b)\\
0, & a \neq \argmax{b} Q(b)
\end{cases}
$$

```python
def greedy(Q):
    return 1 + np.argmax(Q)
```

```python
greedy(Q)
```

But this was driven by purely random action selection.  Instead, we would like to start using the greedy action as we gain confidence in the $Q$ values.  A simple way to do this is to use an $\epsilon$-greedy action choice.

```python
def epsilonGreedy(Q,epsilon):
    if np.random.uniform() < epsilon:
        return 1 + np.random.choice(2) # random action
    else:
        return 1 + np.argmax(Q)        
```

and slowly decrease $\epsilon$ from 1 to 0, asymptotically.  

```python
Q = [0,0]
rho = 0.2
epsilon = 1.0
log = {'Q':[], 'a':[], 'r':[], 'e':[]}
for steps in range(100):
    a = epsilonGreedy(Q,epsilon)
    r = reward(a)
    Q[a-1] = Q[a-1] + rho * (r - Q[a-1])
    epsilon = 0.9 * epsilon
    
    log['Q'].append(Q[:])
    log['a'].append(a)
    log['r'].append(r)
    log['e'].append(epsilon)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(4,1,1)
plt.plot(log['a'],'o-')
plt.ylabel("Action Chosen")

plt.subplot(4,1,2)
plt.plot(log['r'])
plt.ylabel("Reward")

plt.subplot(4,1,3)
plt.plot(log['e'])
plt.ylabel('$\epsilon$')

plt.subplot(4,1,4)
plt.plot(np.array(log['Q']),lw=3)
plt.xlabel("Step")
plt.ylabel("Q value")
plt.legend(('$Q(a=1)$', '$Q(a=2)$'));
```

Notice that now the action chosen quickly converges on action 2, the optimal action.

This did not require very many steps.  How can we make this simple problem more challenging?  What if the reward values for the two actions are more similar?

```python
def reward(a):
    z = np.random.normal(0,30)
    if a == 1:
        return 10 + z
    else:
        return 20 + z
```

```python
Q = [0,0]
rho = 0.2
epsilon = 1.0
log = {'Q':[], 'a':[], 'r':[], 'e':[]}
for steps in range(100):
    a = epsilonGreedy(Q,epsilon)
    r = reward(a)
    Q[a-1] = Q[a-1] + rho * (r - Q[a-1])
    epsilon = 0.9 * epsilon
    
    log['Q'].append(Q[:])
    log['a'].append(a)
    log['r'].append(r)
    log['e'].append(epsilon)
    
plt.figure(figsize=(10,8))

plt.subplot(4,1,1)
plt.plot(log['a'],'o-')
plt.ylabel("Action Chosen")

plt.subplot(4,1,2)
plt.plot(log['r'])
plt.ylabel("Reward")

plt.subplot(4,1,3)
plt.plot(log['e'])
plt.ylabel('$\epsilon$')

plt.subplot(4,1,4)
plt.plot(np.array(log['Q']),lw=3)
plt.xlabel("Step")
plt.ylabel("Q value")
plt.legend(('$Q(a=1)$', '$Q(a=2)$'));
```

With larger noise in the reward function, we converged on the incorrect action.  We can slow the learning down by decreasing $\rho$, or slow the decrease of $\epsilon$.  Let's start by defining a function to repeat the above experiment.

```python
def run1(rho,epsilonDecay,nSteps):
    Q = [0,0]
    epsilon = 1.0
    log = {'Q':[], 'a':[], 'r':[], 'e':[]}
    for steps in range(nSteps):
        a = epsilonGreedy(Q,epsilon)
        r = reward(a)
        Q[a-1] = Q[a-1] + rho * (r - Q[a-1])
        epsilon = epsilonDecay * epsilon
    
        log['Q'].append(Q[:])
        log['a'].append(a)
        log['r'].append(r)
        log['e'].append(epsilon)
    
    plt.figure(figsize=(10,8))

    plt.subplot(4,1,1)
    plt.plot(log['a'],'o-')
    plt.ylabel("Action Chosen")

    plt.subplot(4,1,2)
    plt.plot(log['r'])
    plt.ylabel("Reward")

    plt.subplot(4,1,3)
    plt.plot(log['e'])
    plt.ylabel('$\epsilon$')

    plt.subplot(4,1,4)
    plt.plot(np.array(log['Q']),lw=3)
    plt.xlabel("Step")
    plt.ylabel("Q value")
    plt.legend(('$Q(a=1)$', '$Q(a=2)$'));
```

```python
run1(0.05,0.99, 300)
```

We are learning an accurate estimate of $Q(a=2)$, but not $Q(a=1)$.  Let's learn for more steps and decrease $\epsilon$ slower.

```python
run1(0.1,0.995, 400)
```

## Two states, two actions

Your world may be in two different states.  Each state has their own probabilities, so we need to keep track of separate $Q$ values in each state.  Since we still have two actions, $Q$ will now contain four values, one for each state, $s$, and action, $a$, or $Q(s,a)$.

Let's assume our world chooses which state we are in at random.   Let's also define our reward function so that Action 1 is preferred (has a higher reward) in State 1 and Action 2 is preferred in State 2.

```python
def reward(s,a):
    z = np.random.normal(0,10)
    if s == 1:
        if a == 1:
            return 20 + z
        else:
            return 10 + z
    else:
        if a == 1:
            return -5 + z
        else:
            return 20 + z
```

This can be defined more concisely using a reward matrix.

```python
def reward(s,a):
    R = np.array([[20,10],[-5,10]])
    z = np.random.normal(0,2)
    return R[s-1,a-1] + z
```

```python
def epsilonGreedy(Q,state,epsilon):
    if np.random.uniform() < epsilon:
        return np.random.choice((1,2))
    else:
        return 1 + np.argmax(Q[state-1,:])   
```

Now, we run a similar experiment.  How $Q$ is a 2x2 matrix, with the state indexing the row and the action indexing the column.

```python
def run2(rho,epsilonDecay,nSteps):
    Q = np.array([[0,0],[0,0]])
    epsilon = 1.0
    log = {'Q':[], 's':[], 'a':[], 'r':[], 'e':[]}
    for steps in range(nSteps):
        s = 1 + np.random.choice(2)
        a = epsilonGreedy(Q,s,epsilon)
        r = reward(s,a)
        Q[s-1,a-1] = Q[s-1,a-1] + rho * (r - Q[s-1,a-1])
        epsilon = epsilonDecay * epsilon
    
        log['Q'].append(Q.flatten())
        log['s'].append(s)
        log['a'].append(a)
        log['r'].append(r)
        log['e'].append(epsilon)
    
    plt.figure(figsize=(10,8))

    plt.subplot(5,1,1)
    plt.plot(log['s'],'o-')
    plt.ylabel('state')
    
    plt.subplot(5,1,2)
    plt.plot(log['a'],'o-')
    plt.ylabel("Action Chosen")

    plt.subplot(5,1,3)
    plt.plot(log['r'])
    plt.ylabel("Reward")

    plt.subplot(5,1,4)
    plt.plot(log['e'])
    plt.ylabel('$\epsilon$')

    plt.subplot(5,1,5)
    
    plt.plot(log['Q'],lw=2)
    plt.xlabel("Step")
    plt.ylabel("Q value")
    plt.legend(('$Q(s=1,a=1)$', '$Q(s=1,a=2)$', '$Q(s=2,a=1)$', '$Q(s=2,a=2)$'));
    
    return log
```

```python
g = run2(0.1,0.9,100)
```

## Ten states, three actions

Okay. That works for two states.  What if we have a chain of 11 states arranged in a row and action 1 takes us one state to the right and action -1 takes us one state to the left, and action 0 leaves us in same state.  We remain in the right-most and left-most states if we try to move right, or left, respectively.

And, and this is a BIG AND, what if we don't get a reward until we land in the middle state!  The reward is **delayed** for many of the actions.

One thing at a time. Let's define the state transition function.

$$ f(s_t,a_t) = \max( \min(s_t+a_t, 10), 0)$$

```python
def f(s,a):
    return max(min(s+a, 10), 0)
print f(5,-1), f(5,0), f(5,1), f(0,-1), f(0,1), f(10,-1), f(10,1)
```

Need to redefine epsilonGreedy for the different set of possible actions.

```python
def epsilonGreedy(Q,state,epsilon):
    if np.random.uniform() < epsilon:
        return np.random.choice((-1,0,1)) # random action
    else:
        return -1 + np.argmax(Q[state,:])   
```

Now for the reward function.

$$r(s_t,a_t) = \begin{cases}
1, \text{ if } s_t = 5,\\
0, \text{ otherwise }
\end{cases}$$

```python
def reward(s,a):
    if s == 5:
        return 1
    else:
        return 0
print reward(3,1), reward(5,1), reward(5,2)
```

But how do we learn optimal actions for states that receive no reward?

Here's the key to the power of reinforcement learning.  We design algorithms to optimize the sum of future rewards, not just the immediate reward. (Just like in dynamic programming.)  Our greedy policy can still be 

$$\pi(s) = \argmax{a\in \{1,2\}} Q(s,a),$$

if we redefine $Q$ to be

$$Q(s_t,a_t) \approx \sum_{k=0}^\infty r_{t+k}$$

Of course, this sum is undefined, so we do the usual trick of weighting each term by an exponentially decreasing discount factor.

$$Q(s_t,a_t) \approx \sum_{k=0}^\infty \gamma^k r_{t+k}\;\;,\text{ for } 0 < \gamma \le 1.$$

We cannot directly use the sum as a target in a squared error minimization problem, because of the $\infty$.  Instead, we make the huge assumption that we can use $Q(s_{t+1},a_{t+1})$ as a pretty good estimation of $\sum_{k=1}^\infty \gamma^k r_{t+k}$ to which we only need to add $r_t$ to get the above right-hand side.

$$\begin{align*}
\sum_{k=0}^\infty \gamma^k r_{t+k} &= r_t + \sum_{k=1}^\infty \gamma^k r_{t+k}\\
&\approx r_t + \gamma Q(s_{t+1},a_{t+1})
\end{align*}$$

Now we can form the optimization problem

$$\text{Minimize } (r_t + \gamma Q(s_{t+1},a_{t+1}) - Q(s_t,a_t))^2$$

and go down the gradient of this with respect to the $Q$ values.

$$ Q(s_t,a_t) = Q(s_t,a_t) + \rho (r_{t+1} +\gamma Q(s_{t+1},a_{t+1}) - Q(a_t))$$








```python
def run11(rho,epsilonDecay,gamma,nSteps):
    Q = np.zeros((11,3))
    epsilon = 1.0
    log = {'Q':[], 's':[], 'a':[], 'r':[], 'e':[]}
    s = 5 # initial state
    a = epsilonGreedy(Q,s,epsilon)
    for steps in range(nSteps):
        sNext = f(s,a)
        r = reward(sNext,a)
        epsilon = epsilonDecay * epsilon
        aNext = epsilonGreedy(Q,sNext,epsilon)
        Q[s,a+1] = Q[s,a+1] + rho * (r + gamma * Q[sNext,aNext+1] - Q[s,a+1])

        s = sNext
        a = aNext
        
        log['Q'].append(Q.flatten())
        log['s'].append(s)
        log['a'].append(a)
        log['r'].append(r)
        log['e'].append(epsilon)
    
    plt.figure(figsize=(10,8))

    plt.subplot(5,1,1)
    plt.plot(log['s'],'.-')
    plt.ylabel('state')
    
    plt.subplot(5,1,2)
    plt.plot(log['a'],'o-')
    plt.ylabel("Action Chosen")

    plt.subplot(5,1,3)
    rs = np.array(log['r'])
    plt.plot(rs)
    plt.plot(np.arange(10,len(rs)+1,10),rs.reshape((-1,10)).mean(1),lw=3,alpha=0.8)
    plt.ylabel("Reward")

    plt.subplot(5,1,4)
    plt.plot(log['e'])
    plt.ylabel('$\epsilon$')

    plt.subplot(5,1,5)
    
    plt.plot(log['Q'],lw=2)
    plt.xlabel("Step")
    plt.ylabel("Q value")
    
    return log
```

```python
g = run11(0.01,0.995,0.9,1000)
```

## Continuous state, two actions

The discrete representation of state works okay until the number of possible states is too large.  The number of states becomes large when the state has multiple dimensions.  If the state is two dimensional and we have 10 values per dimension, we end up with 10x10=100 different states.  With 10 values per dimension, and $n$ dimensions, we will have $10^n$ states.  The number of states grows exponentially with the number of dimensions.

For real robot applications, we have have 10's of dimensions, one for each sensor measurement.  We need a better way to represent the $Q$ function other than a discretely-valued table.  We need a continuous function approximator for $Q(s,a)$.

How about a neural network?  With $n$ sensor measurements as input, plus an input for the single action, the neural network will have $n+1$ inputs.  It will have a single output, $Q(s,a)$, where $s$ is an $n$-dimensional vector, and $a$ is a scalar.

Let's define a simple control problem.  We have a marble that can move left and right on a track with limits at $x=0$ and $x=10$.  The marble has dynamics, meaning its state includes $x$ and $\dot{x}$.  The three available actions are $a=-1$, $a=0$, and $a=1$, which are forces in the negative and positive directions, respectively.  

The dynamics of the marble are defined by
$$\begin{align*}
x_{t+1} &= x_t + \Delta\;\dot{x}_t\\
\dot{x}_{t+1} & = \dot{x}_t + \Delta\; \ddot{x}_t
\end{align*}$$
where
$$\ddot{x}_t = F\;a_t$$
and $F=10$ Nt and $\Delta$ is a small increment of time, like $\Delta = 0.01$ s.

```python
def nextState(s,a,delta,F):
    x,xdot = s
    return [x + delta*xdot, xdot + delta * F * a]

nextState( (3,2), 1, 0.1, 10 ),  nextState((3,2), -1, 0.1, 10), nextState((3,-1), -0, 0.1, 10)
```

Let's define the objective of the marble control problem to be to keep the marble between $x = 4$ and $x=6$.

```python
def reward(s,a):
    return 1 if 4 < s[0] < 6 else 0
reward((1,0),1), reward((5,1),-1)
```

Since $Q$ is estimated by a function now and not just a value in a table, let's modify the __epsilonGreedy__ function to return an action and the computed $Q$ value.

```python
def epsilonGreedy(Qnet,s,possibleActions,epsilon):
    Qs = Qnet.use(np.array([s + [a] for a in possibleActions]))
    if np.random.uniform < epsilon:
        a = np.random.choice(possibleActions)
        Q = Qs[a+1]
    else:
        a = possibleActions[np.argmax(Qs)]
        Q = Qs[a+1]
    return a,Q
```

Let's use the Scaled Conjugate Gradient algorithm to train our neural network to approximate the $Q$ function.  To use it, we must collect a batch of samples, each one a tuple of state, action, reward, next state, next action.

```python
import random
def getSamples(net,possibleActions,numSamples,epsilon,epsilonDecay,deltaT,force):
    X = np.zeros((numSamples,net.getSize()[0]))
    R = np.zeros((numSamples,1))
    Qn = np.zeros((numSamples,1))
    Q = np.zeros((numSamples,1))

    # Initial state, action, and Q value
    s = [random.randint(0,10), 0.0]  # x random between 0 and 10.  xdot is 0
    a,q = epsilonGreedy(net,s,possibleActions,epsilon)

    # Collect data from numSamples steps
    for step in xrange(numSamples):
        sn = nextState(s,a,deltaT,force)        # Update state, sn from s and a
        rn = reward(sn,a)   # Calculate resulting reinforcement
        an,qn = epsilonGreedy(net,sn,possibleActions,epsilon) # Forward pass for time t+1

        X[step,:], R[step,0], Qn[step,0], Q[step,0] = s+[a], rn, qn, q

        s,a,q = sn,an,qn               # advance one time step
        epsilon *= epsilonDecay         # decay exploration probability
        
    return (X,R,Qn,Q,epsilon)
```

Using the code in <a href="http://www.cs.colostate.edu/~anderson/cs645notebooks/neuralnetworks.tar">neuralnetworks.tar</a>, we initialize the $Q$ function as

```python
import neuralnetworks as nn
import mlutils as ml
ni = 2+1  # state as 2 components, x and xdot
nh = 10
no = 1
Qnet = nn.NeuralNetworkQ(ni,nh,no, ((0,10),(-3,3),(-1,1)))
```

```python
def runMarble(epsilonDecay,gamma,nQTrain, nStepsPerQTrain, maxSCGIterations):
    deltaT = 0.1
    force = 10
    ni = 2+1  # state as 2 components, x and xdot
    nh = 10
    no = 1
    Qnet = nn.NeuralNetworkQ(ni,nh,no, ((0,10),(-3,3),(-1,1)))
    possibleActions = (-1,0,1)
    epsilon = 1.0
    log = {'Q':[], 's':[], 'a':[], 'r':[], 'e':[]}
    for ntrains in range(nQTrain):
        # Collect nStepsPerRep samples of X, R, Qn, and Q, and update epsilon
        X,R,Qn,Q,epsilon = getSamples(Qnet,possibleActions,nStepsPerQTrain,epsilon,epsilonDecay,deltaT,force)
        # Update the Q neural network.
        nIterations = max(5,maxSCGIterations * epsilon)
        Qnet.train(X,R,Qn,Q,gamma=gamma, nIterations=nIterations,
              weightPrecision=1e-8, errorPrecision=1e-10)
        # Rest is for plotting
        if ntrains == 0:
            log['Q'] = Q
            log['s'] = X[:,0:2]
            log['a'] = X[:,2:3]
        else:
            log['Q'] = np.vstack((log['Q'], Q))
            log['s'] = np.vstack((log['s'],X[:,0:2]))
            log['a'] = np.vstack((log['a'],X[:,2:3]))              
            
        log['r'].append(np.mean(R))
        log['e'].append(epsilon)
        
    return X,Qnet,log
```

```python
X,Qnet,log = runMarble(0.9999, 0.95, 1000,1000, 50)
```
